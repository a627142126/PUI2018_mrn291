### 2.1 Get Data with the American Fact Finder API

```python
#read in in the variables available. the info you need is in the 1 year ACS data
url = "https://api.census.gov/data/2016/acs/acs1/variables.json"
resp = requests.request('GET', url)
aff1y = json.loads(resp.text)
```

```python
#keyword for the number of households
keyNhouseholds = 'B28002_001E'
aff1y['variables'][keyNhouseholds]
```

```python
#keyword for the number of households with broadband
keyBBhouseholds = 'B28002_004E'
aff1y['variables'][keyBBhouseholds]
```

```python
# getting the broadband access number of households
url = "https://api.census.gov/data/2016/acs/acs1?get=" + keyBBhouseholds +\
",NAME&for=public%20use%20microdata%20area:*&in=state:36&key=" + myAPI
resp = requests.request('GET', url).content
pumaBB = pd.read_csv(io.StringIO(resp.decode('utf-8').replace('[','').replace(']','')))

pumaBB.head()
```

```python
#getting the first row of the table: the total number of households
url = "https://api.census.gov/data/2016/acs/acs1?get=" + keyNhouseholds +\
",NAME&for=public%20use%20microdata%20area:*&in=state:36&key=" + myAPI
resp = requests.request('GET', url).content
pumaHH = pd.read_csv(io.StringIO(resp.decode('utf-8').replace('[','').replace(']','')))

pumaHH.head()
```

### 2.2 Generate a feature for the percentage of households with broadband access.

```python
puma_total = pd.merge(pumaHH, pumaBB, on=['state', 'public use microdata area', 'NAME'])
```

```python
puma_total.drop(labels=['Unnamed: 4_x', 'Unnamed: 4_y'], axis=1, inplace=True)
puma_total.rename(columns={'B28002_001E':'Total_HH', 'B28002_004E':'BB_HH', 'public use microdata area':'puma'}, inplace=True)
puma_total['BB_Percent'] = (puma_total['BB_HH'] / puma_total['Total_HH'].astype(float) * 100)
```

### 3.1 Link Data

```python
link_data = pd.read_csv("https://raw.githubusercontent.com/fedhere/PUI2018_fb55/master/HW7_fb55/linkNYClocations.csv", index_col=0)
link_data.head()
```

```python
# combine lat and lon from linkNYC to one column
link_data['latlong'] = list(zip(link_data['longitude'], link_data['latitude']))

# Create Point Geometry based on lonlat column
link_data['geometry'] = link_data[['latlong']].applymap(lambda x: shapely.geometry.Point(x))
link_data.head()
```

```python
#turn it into a geopandas dataframe
link_data = gpd.GeoDataFrame(link_data)
type(link_data)
```

```python
link_data.crs = from_epsg(4326)
```

### Plot the location of the LinkNYC stations on top of a choropleth of broadband access percentage in 5 equal intervals.

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.set_aspect('equal')
ax.axis('off')
plt.title("Percent of Households with Broadband Access by NYC PUMAs")
puma_broadband.to_crs(naec).plot(ax=ax, cmap='Blues', scheme = 'equal_interval', k=5, column='BB_Percent', edgecolor='black', lw=2, legend=True)
#im = ax.imshow(puma_broadband['BB_Percent'], cmap='viridis')
#cax = divider.append_axes('right', size='5%', pad=0.05)
#cbar = fig.colorbar(im, ax=ax)
link_data.to_crs(naec).plot(ax=ax, marker='o', c="date_link_", markersize=6, cmap="Oranges")
plt.show()
```

Figure showing the location of LinkNYC stations overlaid on a choropleth map of percentage of households with broadband access by public use microdata areas.

### 5 Find the number of linkNYC locations per person by PUMA
#### 5.1 with the AFF API from ACS get the total population by puma and merge it into your GeoDF

```python
key_total_pop = 'B00001_001E'
url = "https://api.census.gov/data/2016/acs/acs1?get=" + key_total_pop +\
",NAME&for=public%20use%20microdata%20area:*&in=state:36&key=" + myAPI
resp = requests.request('GET', url).content
puma_pop = pd.read_csv(io.StringIO(resp.decode('utf-8').replace('[','').replace(']','')))
puma_pop.head()
```

```python
puma_pop.rename(columns = {'public use microdata area':'puma', 'B00001_001E':'total_pop'}, inplace=True)
puma_pop.drop('Unnamed: 4', axis=1, inplace=True)
```

```python
puma_pop.dtypes
```

```python
#merge population with puma_broadband
puma_pop_bb = pd.merge(puma_pop, puma_broadband, on='puma')
puma_pop_bb = gpd.GeoDataFrame(puma_pop_bb)
puma_pop_bb.head()
```

```python
#spatial join to link_data dataframe
#first check crs
puma_pop_bb.crs = from_epsg(4326)
print(puma_pop_bb.crs, link_data.crs)
```

```python
#do spatial join
link_puma = gpd.sjoin(link_data, puma_pop_bb)[["puma", "link_site"]].groupby("puma").count().reset_index()
link_puma.head()
```

```python
#join link counts back into puma_pop_bb
puma_pop_bb_link = pd.merge(left = puma_pop_bb, right = link_puma, on='puma', how='left')
puma_pop_bb_link.shape
```

```python
puma_pop_bb_link.columns
```

```python
puma_pop_bb_link['link_per_100'] = puma_pop_bb_link['link_site'] / (puma_pop_bb_link['total_pop']/100)
```

```python
puma_pop_bb_link['link_per_100'].describe()
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.set_aspect('equal')
ax.axis('off')
plt.title("Percent of Households with Broadband Access by NYC PUMAs")
puma_pop_bb_link.to_crs(naec).plot(ax=ax, cmap='gray', scheme = 'equal_interval', k=10, column='link_per_100', edgecolor='black', lw=2, legend=True)
#im = ax.imshow(puma_broadband['BB_Percent'], cmap='viridis')
#cax = divider.append_axes('right', size='5%', pad=0.05)
#cbar = fig.colorbar(im, ax=ax)
link_data.to_crs(naec).plot(ax=ax, marker='o', c="date_link_", markersize=6, cmap="Oranges")
plt.show()
```

Figure showing the location of Link NYC stations overlaid on a choropleth map of the number of link stations per 100 residents by public use microdistrict area.
